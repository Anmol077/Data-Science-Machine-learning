```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [4,5,2]
plt.plot(x,y)
```

```python
x=[1,2,3]
y=[4,5,2]
plt.plot(x,y)
plt.title('Line Graph')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
x1=[1,2,3]
y1=[4,5,2]
x2=[6,7,8]
y2=[9,10,7]
plt.plot(x1,y1,label='first',color='r',linewidth=5)
plt.plot(x2,y2,label='first',color='m',linewidth=4)
plt.title('style graph')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.grid(True,color='lightblue')
plt.show()
```

```python
x1=[1,3,5,7]
y1=[5,2,6,6]
x2=[2,4,6,8]
y2=[6,3,5,2]
plt.bar(x1,y1,label='first')
plt.bar(x2,y2,label='Second')
plt.title('Bar graph')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.legend()
plt.show()
```

```python
sale_values=[12,10,5,7,0,14,15,24,10,11,15,30,18,11,10,18,30]
bins = [0,5,10,15,20,25,30]
plt.hist(sale_values,bins,histtype='step',rwidth=0.6)
plt.title('Histogram graph')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
sale_values=[12,10,5,7,0,14,15,24,10,11,15,30,18,11,10,18,30]
bins = [0,5,10,15,20,25,30]
plt.hist(sale_values,bins,histtype='bar',rwidth=0.6)
plt.title('Histogram graph')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```

```python
x=[1,3,4,8,5,6,7]
y=[3,5,6,7,8,9,3]
plt.scatter(x,y,color='b')
plt.xlabel('x axis')
plt.ylabel('y,axis')
plt.show()
```
